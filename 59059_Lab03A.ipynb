```python
#Import libraries 
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Get directory and read file 
cwd = os.getcwd()
os.chdir(path='/Users/lukegorman/Desktop/Programming for DS')

#read each sheet
price = pd.read_excel('rentingMedia.xlsx', sheet_name='price')
rent = pd.read_excel('rentingMedia.xlsx', sheet_name='rent')
rentab = pd.read_excel('rentingMedia.xlsx', sheet_name='rentab')
```

```python
#Create dataframes for each sheet
price = pd.DataFrame(price)
rent = pd.DataFrame(rent)
rentab = pd.DataFrame(rentab)
```

```python
# Transposing the data and resetting the index.
rent = rent.set_index('price').T.reset_index()
rent['date']=pd.to_datetime(rent['index'], format='%d%m%Y').dt.date
rent =rent.set_index(['date']).drop('index',axis=1)
rent.head()
```

```python

#transposing the data and resetting the index for price
price = price.set_index('price').T.reset_index()
price['date']=pd.to_datetime(price['index'], format='%d%m%Y').dt.date
price =price.set_index(['date']).drop('index',axis=1)
price.head()

```

```python
#transposing the data and resetting the index for rentab
rentab = rentab.set_index('price').T.reset_index()
rentab['date']=pd.to_datetime(rentab['index'], format='%d%m%Y').dt.date
rentab =rentab.set_index(['date']).drop('index',axis=1)
rentab.head()
```

```python
sns.set_theme(style= 'whitegrid')
sns.lineplot(data=price, palette='tab20b', linewidth=1.5, legend= False, dashes=False)
```

```python
ts=price[['Arroios', 'Ajuda','Benfica']].plot(subplots=True,color='green',figsize=(8, 6))
```

```python
sns.set_theme(style= 'darkgrid')
lg = sns.lineplot(data=rent['Arroios'], palette='tab10', linewidth=1.5,legend= True)
```

```python
ts= rentab[['Ajuda', 'Arroios','Parque das Nações']].plot.bar(figsize=(15, 5))
```

```python
pf = sns.relplot(
    data=price, kind="line",
    x="date", y="Marvila", height=5, aspect=4, linewidth=2.5)
```

```python
# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
pf = sns.heatmap(rent, annot=False, linewidths=.2, ax=ax, xticklabels=True, yticklabels=True)
```
